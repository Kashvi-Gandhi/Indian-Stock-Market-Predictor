```python
import pandas as pd
import numpy as np

# 1. Load the dataset
# Assuming your notebook is in the 'notebooks' folder and data is in the 'data' folder
df = pd.read_csv('../data/raw_data.csv')

# 2. Clean and format the Date column
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df = df.set_index('Date')

# 3. Handle missing values 
# Forward-fill carries the previous day's price forward if a day is missing
df = df.ffill().dropna()

print(f"Dataset loaded successfully! Total rows: {df.shape[0]}")
df.head()
```

```python
# 1. Create Moving Averages (SMA)
# The 50-day and 200-day SMAs are standard indicators used by real traders
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()

# 2. Calculate Daily Returns (Volatility)
df['Daily_Return'] = df['Close'].pct_change()

# 3. DEFINE THE TARGET VARIABLE (Crucial Step)
# We want to predict TOMORROW'S closing price. 
# We do this by taking the 'Close' column and shifting it UP by 1 row.
df['Target_Next_Close'] = df['Close'].shift(-1)

# 4. Drop NaN values created by our new features
# The first 200 rows will be dropped because they don't have enough history for the 200-day SMA.
# The very last row is dropped because we don't know tomorrow's price yet.
df = df.dropna()

print(f"Features created! Remaining rows for training: {df.shape[0]}")
# Let's look at the end of the dataset to see our new columns
df[['Close', 'SMA_50', 'SMA_200', 'Target_Next_Close']].tail()
```

```python
import plotly.graph_objects as go

# Plotting 20 years of daily data is too messy. 
# Let's just plot the last 500 days to clearly see the trends.
plot_df = df.tail(500) 

# Build the Candlestick chart
fig = go.Figure(data=[
    go.Candlestick(
        x=plot_df.index,
        open=plot_df['Open'],
        high=plot_df['High'],
        low=plot_df['Low'],
        close=plot_df['Close'],
        name='Tata Motors'
    ),
    # Add the 50-day Moving Average (Blue line)
    go.Scatter(x=plot_df.index, y=plot_df['SMA_50'], line=dict(color='blue', width=1.5), name='50-Day SMA'),
    
    # Add the 200-day Moving Average (Red line)
    go.Scatter(x=plot_df.index, y=plot_df['SMA_200'], line=dict(color='red', width=1.5), name='200-Day SMA')
])

# Format the layout to look like a pro trading terminal
fig.update_layout(
    title='Tata Motors (TATAMOTORS.NS) - Last 500 Days',
    yaxis_title='Stock Price (INR)',
    xaxis_title='Date',
    template='plotly_dark', # Dark mode!
    xaxis_rangeslider_visible=False # Hides the bulky slider at the bottom
)

fig.show()
```

```python
from sklearn.model_selection import train_test_split

# 1. Define what the model will study (Features)
# We give it the daily prices, volume, and the smart features we created
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'SMA_50', 'SMA_200', 'Daily_Return']

# 2. Define what the model must predict (Target)
target = 'Target_Next_Close'

X = df[features]
y = df[target]

# 3. The Chronological Split (CRITICAL STEP)
# We will use the first 80% of the timeline to train, and the last 20% to test.
# Notice 'shuffle=False'. This ensures the dates stay in order!
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print(f"Training Data: {X_train.shape[0]} days")
print(f"Testing Data: {X_test.shape[0]} days")
print(f"Training starts on: {X_train.index[0].date()} and ends on: {X_train.index[-1].date()}")
print(f"Testing starts on: {X_test.index[0].date()} and ends on: {X_test.index[-1].date()}")
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# 1. Initialize the Scaler
scaler = MinMaxScaler()

# 2. Scale the features
# IMPORTANT: We only 'fit' (learn the min/max) on the TRAINING data to prevent data leakage.
# Then we transform both train and test data.
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 3. Initialize and Train the Random Forest Model
print("Training Random Forest Baseline Model... (This might take 10-20 seconds)")
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train)
print("Training Complete!")

# 4. Make Predictions on the unseen Test data
rf_predictions = rf_model.predict(X_test_scaled)

# 5. Evaluate the Model
rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))
mae = mean_absolute_error(y_test, rf_predictions)

print(f"Baseline Root Mean Squared Error (RMSE): ₹{rmse:.2f}")
print(f"Baseline Mean Absolute Error (MAE): ₹{mae:.2f}")
```

```python
import matplotlib.pyplot as plt

# Create a clean plot comparing Actual vs Predicted prices
plt.figure(figsize=(14, 6))

# Plot the actual test prices
plt.plot(y_test.index, y_test.values, label='Actual Tata Motors Price', color='blue')

# Plot our model's predictions
plt.plot(y_test.index, rf_predictions, label='Random Forest Predictions', color='orange', alpha=0.7)

plt.title('Tata Motors: Actual vs Predicted Prices (Baseline Model)')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 1. THE FIX: Create a dedicated scaler for the Target Variable
target_scaler = MinMaxScaler()

# We must reshape the data slightly for the scaler, then flatten it back
y_train_scaled = target_scaler.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_test_scaled = target_scaler.transform(y_test.values.reshape(-1, 1)).flatten()

# 2. Set the window size
time_steps = 60 

def create_sequences(X_data, y_data, time_steps):
    X_seq, y_seq = [], []
    for i in range(len(X_data) - time_steps):
        X_seq.append(X_data[i : i + time_steps])
        y_seq.append(y_data[i + time_steps]) 
    return np.array(X_seq), np.array(y_seq)

# 3. Create the 3D sequences using the SCALED targets
X_train_lstm, y_train_lstm = create_sequences(X_train_scaled, y_train_scaled, time_steps)
X_test_lstm, y_test_lstm = create_sequences(X_test_scaled, y_test_scaled, time_steps)

print(f"Data Windowing Complete! Target variables are now scaled.")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 1. Initialize the Sequential Model
model = Sequential()

# 2. Add the first LSTM Layer
# input_shape must match our window size and number of features
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
model.add(Dropout(0.2)) # Randomly turns off 20% of neurons to prevent overfitting

# 3. Add a second LSTM Layer
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))

# 4. Add the Output Layer (predicting 1 single price)
model.add(Dense(units=1))

# 5. Compile the Model
# 'adam' is a great default optimizer, 'mse' (Mean Squared Error) is our loss function
model.compile(optimizer='adam', loss='mean_squared_error')

print("Training the Deep Learning LSTM Model... (This will take a minute or two!)")
# 6. Train the Model
# epochs = how many times it reads the whole dataset. batch_size = how many windows it reads at once.
history = model.fit(X_train_lstm, y_train_lstm, epochs=20, batch_size=32, validation_split=0.1, verbose=1)

# 7. SAVE THE MODEL (Crucial for Phase 5)
model.save('../models/lstm_model.keras')
print("Model saved successfully to the /models folder!")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 1. Make Predictions (These will be decimals between 0 and 1)
lstm_predictions_scaled = model.predict(X_test_lstm)

# 2. THE MAGIC TRICK: Convert decimals back to Real Rupees
lstm_predictions = target_scaler.inverse_transform(lstm_predictions_scaled)

# 3. Get the actual real prices for comparison (slice off the first 60 days)
actual_prices = y_test.values[time_steps:]

# 4. Calculate the Error in real Rupees
lstm_rmse = np.sqrt(mean_squared_error(actual_prices, lstm_predictions))
lstm_mae = mean_absolute_error(actual_prices, lstm_predictions)

print(f"LSTM Root Mean Squared Error (RMSE): ₹{lstm_rmse:.2f}")
print(f"LSTM Mean Absolute Error (MAE): ₹{lstm_mae:.2f}")

# 5. Visualize the Results
plt.figure(figsize=(14, 6))

dates_for_plot = y_test.index[time_steps:] 

plt.plot(dates_for_plot, actual_prices, label='Actual Tata Motors Price', color='blue')
plt.plot(dates_for_plot, lstm_predictions, label='LSTM Predictions', color='green', alpha=0.8)

plt.title('Tata Motors: Actual vs Deep Learning (LSTM) Predictions')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input

# 1. Choose the stocks you want to become "Experts" in
stocks_to_train = sorted([f for f in os.listdir("../data") if f.endswith(".csv")])
time_steps = 60

for stock_file in stocks_to_train:
    print(f"\n{'='*40}")
    print(f"Training Expert Model for: {stock_file}")
    print(f"{'='*40}")
    
    # --- A. Load and Clean ---
    df = pd.read_csv(f'../data/{stock_file}')
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values('Date').set_index('Date').ffill().dropna()
    
    # --- B. Feature Engineering ---
    df['SMA_50'] = df['Close'].rolling(window=50).mean()
    df['SMA_200'] = df['Close'].rolling(window=200).mean()
    df['Daily_Return'] = df['Close'].pct_change()
    df['Target_Next_Close'] = df['Close'].shift(-1)
    df = df.dropna()
    
    # --- C. Scaling ---
    features = ['Open', 'High', 'Low', 'Close', 'Volume', 'SMA_50', 'SMA_200', 'Daily_Return']
    X = df[features]
    y = df['Target_Next_Close']
    
    feature_scaler = MinMaxScaler()
    X_scaled = feature_scaler.fit_transform(X)
    
    target_scaler = MinMaxScaler()
    y_scaled = target_scaler.fit_transform(y.values.reshape(-1, 1)).flatten()
    
    # --- D. Windowing ---
    X_seq, y_seq = [], []
    for i in range(len(X_scaled) - time_steps):
        X_seq.append(X_scaled[i : i + time_steps])
        y_seq.append(y_scaled[i + time_steps])
    X_seq, y_seq = np.array(X_seq), np.array(y_seq)
    
    # --- E. Build and Train Model ---
    model = Sequential()
    model.add(Input(shape=(X_seq.shape[1], X_seq.shape[2])))
    model.add(LSTM(units=50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Training for 10 epochs just to save you time right now!
    model.fit(X_seq, y_seq, epochs=10, batch_size=32, validation_split=0.1, verbose=1)
    
    # --- F. Save the dynamically named model ---
    stock_name = stock_file.replace('.csv', '')
    model.save(f'../models/{stock_name}_lstm.keras')
    print(f"✅ Successfully saved {stock_name}_lstm.keras")
```
